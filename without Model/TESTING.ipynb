```python
# Unit Testing

test_cases_unknown = [
    {"input": "What are the admission requirements for the university?", "expected_tag": "admission_undergraduate_procedure"},
    {"input": "How do I apply for a course?", "expected_tag": "admission_undergraduate_procedure"},
    {"input": "What are the available courses in the university?", "expected_tag": "admission_undergraduateprograms_programsoffering"},
    {"input": "What are the tuition fees for international students?", "expected_tag": "admission_undergraduate_tuition_fee"},
    {"input" : "who is the tup president?", "expected_tag": "president"},
    {"input": "who is the dean of cos", "expected_tag":"officials_academic_affairs"},
    "What are the prerequisites for the computer science course?",
    "Can I take a course in a different semester?",
    "How do I enroll in a course with a full capacity?",
    "What are the course materials for the statistics course?",
    "Can I take a course online?",
    "How do I access the university's library?",
    "What are the available student services on campus?",
    "How do I get help with my assignments?",
    "What are the available scholarships for international students?",
    "How do I report an issue with my accommodation?",
    "What are the available extracurricular activities on campus?",
    "How do I join a student club?",
    "What are the campus facilities available for students?",
    "How do I get involved in research projects?",
    "What are the available internship opportunities?",
    "What are the university's policies on plagiarism?",
    "How do I request a leave of absence?",
    "What are the available student health services?",
    "How do I get a replacement for a lost ID card?",
    "What are the available career counseling services?"
]
```

```python
import json

# Load intents data
with open('../intents.json', 'r') as file:
    data = json.load(file)

intents = data['intents']
```

# Spacy Similarity Testing

```python
import spacy

nlp = spacy.load("en_core_web_md")
```

```python
spacy_results = []
jaccard_results = []
```

```python
def spacy_similarity(user):
    user = nlp(user)
    similarities = []
    tags = []
    for intent in intents:
        for sents in intent['patterns']:
            pattern = nlp(sents)
            # print(user, " <-> ", pattern, user.similarity(pattern))
            similarities.append(user.similarity(pattern))
            tags.append(intent['tag'])
            
    max_similarity = max(similarities)
    max_similarity_index = similarities.index(max_similarity)
    # max_similarity_intent = intents[max_similarity_index]['tag']
    return max_similarity, max_similarity_index, tags

def get_response(index, tags):
    for intent in intents:
        if tags[index] == intent['tag']:
            spacy_results.append(intent['responses'][0])
            print("Response: ", intent['responses'][0])
            break
```

```python
for user_input in test_cases_unknown:
    max_similarity, max_similarity_index, tags = spacy_similarity(user_input)
    print(user_input)
    get_response(max_similarity_index, tags)
    print()
```

# Jaccard Similarity Testing

```python
def jaccard_similarity(user):
    users = nlp(user)
    user = set(user.text.lower() for user in users if not user.is_stop and not user.is_punct)
    similarities = []
    tags = []
    for intent in intents:
        for sents in intent['patterns']:
            patterns = nlp(sents)
            pattern = set(pat.text.lower() for pat in patterns if not pat.is_stop and not pat.is_punct)
            # print(user, " <-> ", pattern, len(user.intersection(pattern)) / len(user.union(pattern)))
            similarities.append(len(user.intersection(pattern)) / len(user.union(pattern)))
            tags.append(intent['tag'])
            
    max_similarity = max(similarities)
    max_similarity_index = similarities.index(max_similarity)
    return max_similarity, max_similarity_index, tags

def get_response(index, tags):
    for intent in intents:
        if max_similarity == 0.0:
            jaccard_results.append("Im Sorry, i did not understand what you said")
            print("Im Sorry, i did not understand what you said")
            break
        if tags[index] == intent['tag']:
            jaccard_results.append(intent['responses'][0])
            print("Response: ", intent['responses'][0])
            break
```

```python
for user_input in test_cases_unknown:
    max_similarity, max_similarity_index, tags = jaccard_similarity(user_input)
    print(user_input)
    get_response(max_similarity_index, tags)
    print()
```
